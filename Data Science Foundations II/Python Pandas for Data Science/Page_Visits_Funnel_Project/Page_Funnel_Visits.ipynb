```python
import pandas as pd
```

Import all the files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
                   
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

#### Step 1: Inspect the DataFrames using `print()` and `head()`

```python
print(visits.head(5))
print(cart.head(5))
print(checkout.head(5))
print(purchase.head(5))
```

#### Step 2: Left merging visits and cart

```python
visits_cart = pd.merge(visits, cart, on='user_id', how='left')
print(visits_cart)
```

#### Step 3: How long is `visits_cart`?

```python
visits_cart_length=len(visits_cart)
print(visits_cart_length, 'rows')
```

#### Step 4: How many timestamps are null for `cart_time`?

```python
carttime_null_count=cart['cart_time'].isnull().sum()
print(carttime_null_count, 'rows')
```

#### Step 5: What percentage only visited?

As for mother set used in denominator for percentage calculation, only length of visits dataframe is enough because to add into cart, check out and then purchase, the user must visit the site!

```python
only_visited = visits_cart[visits_cart['cart_time'].isnull()]
percentage_only_visited = len(only_visited)/len(visits) * 100
print(round(percentage_only_visited, 2), '%')
```

So, there are 1652 rows include null value for cart_time column.

#### Step 6: What percentage placed a t-shirt in their cart but did not checkout?

```python
cart_checkout = pd.merge(cart, checkout, on='user_id', how='left')
only_added_to_cart = cart_checkout[cart_checkout['checkout_time'].isnull()]
percentage_only_added_to_cart = len(only_added_to_cart)/len(cart) * 100
print(round(percentage_only_added_to_cart, 2), '%')
```

#### Step 7: Merge it all together

```python
all_data = pd.merge(visits_cart, cart_checkout, on='user_id', how='left')
all_data = pd.merge(all_data, purchase, on='user_id', how='left')
print(all_data)
```

#### Step 8: % of users who got to checkout but did not purchase

```python
checkout_purchase = pd.merge(checkout, purchase, on='user_id', how='left')
only_checkout = checkout_purchase[checkout_purchase['purchase_time'].isnull()]
percentage_only_checkout = len(only_checkout)/len(checkout) * 100
print(round(percentage_only_checkout, 2), '%')
```

#### Step 9: check each part of the funnel, let's print all 3 of them again

```python
print(visits_cart)
print(cart_checkout)
print(checkout_purchase)
```

*The weakest part of the funnel is clearly getting a person who visited the site to add a tshirt to their cart since the percentage of user only visiting the pages but not adding items into cart is highest (82,6%). Once they've added a t-shirt to their cart it is fairly likely they end up purchasing it. A suggestion could be to make the add-to-cart button more prominent on the front page.*


#### Step 10: adding new column

Adding a column that is the difference between purchase_time and visit_time into the giant funnel.

```python
all_data['time_from_visit_to_purchase'] = all_data.purchase_time - all_data.visit_time
print(all_data['time_from_visit_to_purchase'])
```

#### Step 12: average time to purchase

```python
avg_time_purchase = all_data['time_from_visit_to_purchase'].mean()
print(avg_time_purchase)
```

## Conclusion
After analyzing the funnel data for Cool T-Shirts Inc., we found that the largest drop-off happens when users visit the site but do not add a t-shirt to their cart. This suggests that the homepage or product pages might need improvement to better engage visitors and encourage them to start shopping.

The percentage of users who drop off after adding to cart or after starting checkout is much smaller, but still important to monitor. Optimizing these steps could further improve conversions.

Finally, the average time from visit to purchase shows how long it takes users to complete the buying process. This information can help the company decide if they need to speed up checkout or send reminders to users who wait too long.
