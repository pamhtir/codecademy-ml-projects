2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `"Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
# Filtering a dataset by a list of words
def words_filter(data, input_list):
    # The filter will return true if all the words in input_list are in the data
    filter = lambda x: all(word in x for word in input_list)
    # Return the value in DataFrame format. This will help we have an option to filter specific columns out for readable purposes
    return data.loc[data["Question"].apply(filter)]
    


# Testing the filter function
filtered = words_filter(jeopardy_df, ["King", "England"]) 
print(filtered["Question"]) # Return specific column - column "Question"
````

3. Test your original function with a few different sets of words to try to find some ways your function breaks. Edit your function so it is more robust.

   For example, think about capitalization. We probably want to find questions that contain the word `"King"` or `"king"`.
   
   You may also want to check to make sure you don't find rows that contain substrings of your given words. For example, our function found a question that didn't contain the word `"king"`, however it did contain the word `"viking"` &mdash; it found the `"king"` inside `"viking"`. Note that this also comes with some drawbacks &mdash; you would no longer find questions that contained words like `"England's"`.

````python
# Another version of the words_filter()
def words_filter(data, input_list):
    # Lowercase all the words in list and the data frame
    filter = lambda x: all(word.lower() in x.lower() for word in input_list)
    # Return the filtered rows in format of a DataFrame include key words
    return data.loc[data["Question"].apply(filter)]
    # .loc[] works as a Boolean. It only filters the rows that returned True as for arguement inside the brackets '[]'
    # data["Question"] is actually 'x' indicated in the assignment of 'filter' lambda above. so all the rows of column "Question" will be scanned if some of them contain keywords in 'input_list'
    
# Testing the upgraded filter function
filtered = words_filter(jeopardy_df, ["King", "England"])
print(filtered["Question"]) 
# Again, since filtered is a DataFrame created by using words_filter(), to access one specific column, it should be written as the above syntax
````

4. We may want to eventually compute aggregate statistics, like `.mean()` on the `"Value"` column. But right now, the values in that column are strings. Convert the`"Value"` column to floats. If you'd like to, you can create a new column with float values.

   While most of the values in the `"Value"` column represent a dollar amount as a string, note that some do not &mdash; these values will need to be handled differently!

   Now that you can filter the dataset of question, use your new column that contains the float values of each question to find the "difficulty" of certain topics. For example, what is the average value of questions that contain the word `"King"`?

````python
print(jeopardy_df["Value"].head(50))
````

By examining the column 'Value', '$', a special symbol, is always added in the front of all money amount, which will raise errors if the string executed calculations using built-in functions e.g. mean(), so it should be removed.   
Additionally, it is observed that for each step of 1000, one comma (,) will be added to separate units for readable purposes. However, to execute calculations using math operators, we have to create a new column which include neither such special symbols nor commas.  
Lastly, some rows don't include any valuable strings but string "no value". A number '0' should replace that string.  
New column and calculating mean values for specific columns are done as follows.

````python
# Since special symbol '$' is always added in the front of money amount, we just creata a new column by extracting string from index 1 onwards.
# As for removing commas, it is needed to use .replace() function
# The following line will create a new column which satisfy the two conditions above
jeopardy_df["Refined Value"] = jeopardy_df["Value"].apply(lambda x: float(x[1:].replace(',', '')) if x != "no value" else 0)
# 'x' is each value stored in each row of column 'Value'     

# Filtering the dataset with the keyword "King" in the column "Question"
filtered = words_filter(jeopardy_df, ["King"])
# Then, finding the average value of the above filtered DataFrame
print(round(filtered["Refined Value"].mean(), 2)) # With being rounded to 2 decimal numbers
````
