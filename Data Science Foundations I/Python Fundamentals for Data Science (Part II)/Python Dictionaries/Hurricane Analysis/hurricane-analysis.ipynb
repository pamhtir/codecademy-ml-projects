   Write a function that returns a new list of updated damages where the recorded data is converted to float values and the missing data is retained as `"Damages not recorded"`.

````python
# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M',
          '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M',
          '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded',
          '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B',
          '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B',
          '91.6B', '25.1B']

# 1
# Update Recorded Damages
conversion = {"M": 1000000,
             "B": 1000000000}

# test function by updating damages
def updating_damages(damages, conversion):
    for i in range(len(damages)):
        if damages[i] == "Damages not recorded":
            damages[i] = "Damages not recorded"
        else:
            if damages[i][-1] == "M":
                damages[i] = float(damages[i][:-1])*conversion["M"]
            elif damages[i][-1] == "B":
                damages[i] = float(damages[i][:-1])*conversion["B"]
                
    return damages

print(updating_damages(damages, conversion))
````

4. You believe that knowing how often each of the areas of the Atlantic are affected by these strong hurricanes is important for making preparations for future hurricanes.

   Write a function that counts how often each area is listed as an affected area of a hurricane. Store and return the results in a dictionary where the keys are the affected areas and the values are counts of how many times the areas were affected.
   
   Test your function on your hurricane dictionary.

````python
# 4
def frequency_organized_affected_areas(all_areas, specific_area = None):

    frequencies = {}
    for i1 in range(len(all_areas)):
        for i2 in range(len(all_areas[i1])):
            if frequencies.get(all_areas[i1][i2]) is None:
                frequencies[all_areas[i1][i2]] = 1
            else:
                frequencies[all_areas[i1][i2]] += 1
    
    if specific_area == None:
        return frequencies
    else:
        return f"{specific_area}: {frequencies[specific_area]}"

print(frequency_organized_affected_areas(areas_affected))
print(f"\n{frequency_organized_affected_areas(areas_affected, specific_area = "Texas")}")
````

7. Just as hurricanes are rated by their windspeed, you want to try rating hurricanes based on other metrics.

   Write a function that rates hurricanes on a mortality scale according to the following ratings, where the key is the rating and the value is the upper bound of deaths for that rating.
   
   ```py
   mortality_scale = {0: 0,
   1: 100,
   2: 500,
   3: 1000,
   4: 10000}
   ```
   
   For example, a hurricane with a `1` mortality rating would have resulted in greater than `0` but less than or equal to `100` deaths. A hurricane with a `5` mortality would have resulted in greater than `10000` deaths.
   
   Store the hurricanes in a new dictionary where the keys are the mortaility ratings and the values are lists containing a dictionary for each hurricane that falls into that mortality rating.
   
   Test your function on your hurricane dictionary.

````python
# 7
# Rating Hurricanes by Mortality
# categorize hurricanes in new dictionary with mortality severity as key
from collections import defaultdict # A module helps avoid state more conditions to test if a key existed in the dictionary yet, making the code less complicated. 

def mortality_rating(all_hurricanes, scale = None):
    rating = defaultdict(list) # defaultdict(list) works by creating an empty list for any missing key of the dictionary made
    for hurricane in all_hurricanes:
        if hurricane["Deaths"] == 0:
            rating["Scale 0"].append(hurricane)
        elif hurricane["Deaths"] > 0 and hurricane["Deaths"] <= 100:
            rating["Scale 1"].append(hurricane)
        elif hurricane["Deaths"] > 100 and hurricane["Deaths"] <= 500:
            rating["Scale 2"].append(hurricane)
        elif hurricane["Deaths"] > 500 and hurricane["Deaths"] <= 1000:
            rating["Scale 3"].append(hurricane)
        elif hurricane["Deaths"] > 1_000 and hurricane["Deaths"] <= 10_000:
            rating["Scale 4"].append(hurricane)
        elif hurricane["Deaths"] > 10_000:
            rating["Scale 5"].append(hurricane)
    
    if scale == None:
        return dict(rating) # Convert rating back to a normal dictionary
    else:
        return f"'{scale}': {rating[scale]}"

        


print(mortality_rating(all_hurricanes))
print(f"\n{mortality_rating(all_hurricanes, scale = "Scale 3")}")

````

9. Lastly, you want to rate hurricanes according to how much damage they cause.

   Write a function that rates hurricanes on a damage scale according to the following ratings, where the key is the rating and the value is the upper bound of damage for that rating.
   ```py
   damage_scale = {0: 0,
   1: 100000000,
   2: 1000000000,
   3: 10000000000,
   4: 50000000000}
   ```
   
   For example, a hurricane with a `1` damage rating would have resulted in damages greater than `0` USD but less than or equal to `100000000` USD. A hurricane with a `5` damage rating would have resulted in damages greater than `50000000000` USD (talk about a lot of money).
   
   Store the hurricanes in a new dictionary where the keys are damage ratings and the values are lists containing a dictionary for each hurricane that falls into that damage rating.
   
   Test your function on your hurricane dictionary.

````python
# 9


from collections import defaultdict

def damaging_rating(all_hurricanes, damages, scale=None):
    rating = defaultdict(list)

    for hurricane, damaging_value in zip(all_hurricanes, damages):
        # skip non-numeric or missing values safely
        try:
            dv = float(damaging_value)
        except (TypeError, ValueError): # Errors can be raised when the float() tries to make strings become float numbers, which is impossible.
            continue 

        if dv == 0:
            rating[0].append(hurricane)
        elif dv <= 100_000_000:
            rating[1].append(hurricane)
        elif dv <= 1_000_000_000:
            rating[2].append(hurricane)
        elif dv <= 10_000_000_000:
            rating[3].append(hurricane)
        elif dv <= 50_000_000_000:
            rating[4].append(hurricane)
        else:
            rating[5].append(hurricane)

    if scale is None:
        return dict(rating)  # normal dict (no 'defaultdict(<class...' print)
    else:
        return f"{scale}: {rating.get(scale, [])}" # A empty list [] will be replaced for a null value if a null value exists


print(damaging_rating(all_hurricanes, new_damages))
print("\n", damaging_rating(all_hurricanes, new_damages, scale = 4))
````
